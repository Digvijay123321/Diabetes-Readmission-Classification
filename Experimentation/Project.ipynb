```python
class NaiveBayes:
    def __init__(self):
        self.class_probs = None
        self.feature_probs = None
    
    def fit(self, X, y):
        # Calculate class probabilities
        classes = np.unique(y)
        self.class_probs = {}
        for c in classes:
            self.class_probs[c] = np.sum(y == c) / len(y)
        
        # Calculate feature probabilities per class
        self.feature_probs = {}
        for c in classes:
            X_c = X[y == c]
            feature_probs_c = {}
            for i in range(X.shape[1]):
                feature_probs_c[i] = {}
                feature_values = np.unique(X[:, i])
                for value in feature_values:
                    feature_probs_c[i][value] = np.sum(X_c[:, i] == value) / len(X_c)
            self.feature_probs[c] = feature_probs_c

    def predict(self, X):
        predictions = []
        for row in X:
            row_probs = {}
            for c in self.class_probs:
                prob = np.log(self.class_probs[c])
                for i in range(len(row)):
                    prob += np.log(self.feature_probs[c][i].get(row[i], 1e-5))  # Add smoothing
                row_probs[c] = prob
            predictions.append(max(row_probs, key=row_probs.get))
        return np.array(predictions)
    
    def predict_probability(self, X):
        """
        Predict probabilities for each class for the input data.
        
        Parameters:
            X: Features for prediction (numpy array).
        
        Returns:
            probabilities: Array of probabilities for each class for each row in X.
        """
        probabilities = []
        for row in X:
            row_probs = {}
            total_prob = 0
            # Calculate unnormalized probabilities for each class
            for c in self.class_probs:
                prob = np.log(self.class_probs[c])
                for i in range(len(row)):
                    prob += np.log(self.feature_probs[c][i].get(row[i], 1e-5))  # Add smoothing
                row_probs[c] = np.exp(prob)  # Convert log-prob back to probability
                total_prob += row_probs[c]
            
            # Normalize to get proper probabilities
            for c in row_probs:
                row_probs[c] /= total_prob
            
            probabilities.append([row_probs[c] for c in sorted(self.class_probs)])
        
        return np.array(probabilities)
```

```python
class LogisticRegression:
    def __init__(self, learning_rate=0.01, iterations=1000):
        self.learning_rate = learning_rate
        self.iterations = iterations
        self.weights = None
        self.bias = 0
    
    def sigmoid(self, z):
        return 1 / (1 + np.exp(-z))
    
    def fit(self, X, y):
        m, n = X.shape
        self.weights = np.zeros(n)
        for _ in range(self.iterations):
            linear_model = np.dot(X, self.weights) + self.bias
            predictions = self.sigmoid(linear_model)
            error = predictions - y
            self.weights -= (self.learning_rate / m) * np.dot(X.T, error)
            self.bias -= (self.learning_rate / m) * np.sum(error)
    
    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        predictions = self.sigmoid(linear_model)
        return np.round(predictions)
    
    def predict_probability(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        return self.sigmoid(linear_model)
```

```python
class SVM:
    def __init__(self, learning_rate=0.01, regularization_param=0.01, iterations=1000):
        self.learning_rate = learning_rate
        self.regularization_param = regularization_param
        self.iterations = iterations
        self.weights = None
        self.bias = 0
    
    def fit(self, X, y):
        m, n = X.shape
        self.weights = np.zeros(n)
        for _ in range(self.iterations):
            for i in range(m):
                if y[i] * (np.dot(X[i], self.weights) + self.bias) < 1:
                    self.weights -= self.learning_rate * (2 * self.regularization_param * self.weights - np.dot(X[i], y[i]))
                    self.bias -= self.learning_rate * y[i]
                else:
                    self.weights -= self.learning_rate * 2 * self.regularization_param * self.weights
        return self
    
    def predict(self, X):
        linear_model = np.dot(X, self.weights) + self.bias
        return np.sign(linear_model)
    
    def predict_probability(self, X):
        """
        Predict probabilities using the sigmoid function.
        
        Parameters:
            X: Features for prediction (numpy array).
        
        Returns:
            probabilities: Probabilities of the positive class (numpy array).
        """
        linear_model = np.dot(X, self.weights) + self.bias
        probabilities = 1 / (1 + np.exp(-linear_model))  # Apply sigmoid
        return probabilities
```

```python
def one_vs_all_custom(model_class, X_train, y_train, X_test):
    """
    Custom implementation of One-vs-All for multi-class classification.
    
    Parameters:
        model_class: A class that implements fit and predict_proba methods.
                     A new instance will be created for each class.
        X_train: Training features (numpy array or similar structure).
        y_train: Training labels (numpy array, one-dimensional).
        X_test: Test features for prediction (numpy array or similar structure).
    
    Returns:
        predictions: Array of predicted class labels for X_test.
        models: List of trained binary classifiers, one for each class.
    """
    classes = np.unique(y_train)  # Get all unique class labels
    models = []  # To store trained models
    scores = np.zeros((X_test.shape[0], len(classes)))  # To store scores for each class

    for i, c in enumerate(classes):
        # Create binary labels for the current class (1 for current class, 0 otherwise)
        binary_y_train = (y_train == c).astype(int)
        
        # Train a new instance of the model for the current class
        model = model_class()
        model.fit(X_train, binary_y_train)
        models.append(model)
        
        # Get probabilities for the current class
        scores[:, i] = model.predict_proba(X_test)
    
    # Final predictions are the classes with the highest score
    predictions = np.argmax(scores, axis=1)
    
    return predictions, models
```
